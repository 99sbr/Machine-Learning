```python
import pandas as pd
print "Reading the csv file..."
review=pd.read_csv('Reviews.csv',delimiter=',')
print "completed !!"
```

```python
Score=review['Score']


def partition(x):
    if x < 3:
        return 'negative'
    return 'positive'

Score = Score.map(partition)
print 'Mapping of score to sentiments done.. !!'
def review_to_words(summary):
        ex=str(summary)
        
        # now we import BeautifulSoup to remove the tags and markups from the text data
        #from bs4 import BeautifulSoup
        #soup=BeautifulSoup(ex)
        clean_text=ex
        # Dealing with Punctuations
        import re
        # Use regular expressions to do a find-and-replace
        letters_only=re.sub("[^a-zA-Z]", " ", clean_text)
        # tokenization
        lower_case=letters_only.lower()
        words=lower_case.split()

        # remove stopword
        import nltk
        from nltk.corpus import stopwords
        stops=set(stopwords.words('english'))
        words= [w for w in words if not w in stops]


        # Stemming
        from nltk.stem.lancaster import LancasterStemmer
        lancaster_stemmer=LancasterStemmer()
        stem_word=[]
        new_set=[]
        for word in words:
            stem_word.append(lancaster_stemmer.stem(word))
            new_set=" ".join( stem_word )
        return new_set
num_reviews = review["Summary"].size    
print "Cleaning and parsing the training set food reviews...\n"
clean_train_reviews = []
for i in range(num_reviews ):
    # If the index is evenly divisible by 1000, print a message
    if( (i+1)%10000 == 0 ):
        print "Review %d of %d\n" % ( i+1, num_reviews )                                                                    
    clean_train_reviews.append( review_to_words( review['Summary'][i] ))
```
