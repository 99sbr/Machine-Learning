```python
import pandas as pd
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")

train_dummy=train.copy()
test_dummy=test.copy()
```

```python
def outlier_detection(x):
    q75, q25 = np.percentile(x, [75 ,25])
    iqr = q75 - q25
    l=q25-iqr
    u=q75+iqr
    return l,u
```

```python

for i in range(0,len(train.iloc[:,0])):
    l,u=outlier_detection(train.iloc[i,:])
    for j in range(0,100):
        if (train.iloc[i,j:j+1]).values[0]>u:
            train.iloc[i,j:j+1].values[0]=u
        elif train.iloc[i,j:j+1].values[0]<l:
            train.iloc[i,j:j+1].values[0]=l
            
for i in range(0,len(test.iloc[:,0])):
    l,u=outlier_detection(test.iloc[i,:])
    for j in range(0,100):
        if (test.iloc[i,j:j+1]).values[0]>u:
            test.iloc[i,j:j+1].values[0]=u
        elif test.iloc[i,j:j+1].values[0]<l:
            test.iloc[i,j:j+1].values[0]=l
```

```python
from xgboost import XGBClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
from sklearn import linear_model
```

```python
model = XGBClassifier()
learning_rate = [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3]
n_estimators=[150,200,250,300,350,400]

param_grid = dict(learning_rate=learning_rate,n_estimators=n_estimators)
kfold = StratifiedKFold(y, n_folds=5, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="accuracy", n_jobs=-1, cv=kfold)
result = grid_search.fit(train,y)
# summarize results
print("Best: %f using %s" % (result.best_score_, result.best_params_))
means, stdevs = [], []
for params, mean_score, scores in result.grid_scores_:
    stdev = scores.std()
    means.append(mean_score)
    stdevs.append(stdev)
    print("%f (%f) with: %r" % (mean_score, stdev, params))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
learning_rate = [0.5,0.2,0.1,0.01,0.001,0.0001]
n_estimators=[100,200,300,400,500,600]

param_grid = dict(learning_rate=learning_rate,n_estimators=n_estimators)
kfold = StratifiedKFold(y, n_folds=5, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="accuracy", n_jobs=-1, cv=kfold)
result = grid_search.fit(train,y)
# summarize results
print("Best: %f using %s" % (result.best_score_, result.best_params_))
means, stdevs = [], []
for params, mean_score, scores in result.grid_scores_:
    stdev = scores.std()
    means.append(mean_score)
    stdevs.append(stdev)
    print("%f (%f) with: %r" % (mean_score, stdev, params))
```
