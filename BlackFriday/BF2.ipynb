```python
import pandas as pd
import numpy as np
train=pd.read_csv('train_mod.csv')
test=pd.read_csv('test_mod.csv')
```

```python
train.head()
```

```python
def getPurchaseVar(compute_df, purchase_df, var_name):
        grouped_df = purchase_df.groupby(var_name)
        min_dict = {}
        max_dict = {}
        mean_dict = {}
        twentyfive_dict = {}
        seventyfive_dict = {}
        for name, group in grouped_df:
                min_dict[name] = min(np.array(group["Purchase"]))
                max_dict[name] = max(np.array(group["Purchase"]))
                mean_dict[name] = np.mean(np.array(group["Purchase"]))
                twentyfive_dict[name] = np.percentile(np.array(group["Purchase"]),25)
                seventyfive_dict[name] = np.percentile(np.array(group["Purchase"]),75)

        min_list = []
        max_list = []
        mean_list = []
        twentyfive_list = []
        seventyfive_list = []
        for index, row in compute_df.iterrows():
                name = row[var_name]
                min_list.append(min_dict.get(name,0))
                max_list.append(max_dict.get(name,0))
                mean_list.append(mean_dict.get(name,0))
                twentyfive_list.append( twentyfive_dict.get(name,0))
                seventyfive_list.append( seventyfive_dict.get(name,0))

        return min_list, max_list, mean_list, twentyfive_list, seventyfive_list
```

```python
min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(train,train,"Product_Category_1")
train['Product_Category_1_min_list']=min_list
train['Product_Category_1_max_list']=max_list
train['Product_Category_1_mean_list']=mean_list
train['Product_Category_1_25_list']=twentyfive_list
train['Product_Category_1_75_list']=seventyfive_list

min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(train,train,"Product_Category_2")
train['Product_Category_2_min_list']=min_list
train['Product_Category_2_max_list']=max_list
train['Product_Category_2_mean_list']=mean_list
train['Product_Category_2_25_list']=twentyfive_list
train['Product_Category_2_75_list']=seventyfive_list

min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(train,train,"Product_Category_3")
train['Product_Category_3_min_list']=min_list
train['Product_Category_3_max_list']=max_list
train['Product_Category_3_mean_list']=mean_list
train['Product_Category_3_25_list']=twentyfive_list
train['Product_Category_3_75_list']=seventyfive_list
```

```python
min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(test,train,"Product_Category_1")
test['Product_Category_1_min_list']=min_list
test['Product_Category_1_max_list']=max_list
test['Product_Category_1_mean_list']=mean_list
test['Product_Category_1_25_list']=twentyfive_list
test['Product_Category_1_75_list']=seventyfive_list

min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(test,train,"Product_Category_2")
test['Product_Category_2_min_list']=min_list
test['Product_Category_2_max_list']=max_list
test['Product_Category_2_mean_list']=mean_list
test['Product_Category_2_25_list']=twentyfive_list
test['Product_Category_2_75_list']=seventyfive_list

min_list,max_list,mean_list,twentyfive_list, seventyfive_list=getPurchaseVar(test,train,"Product_Category_3")
test['Product_Category_3_min_list']=min_list
test['Product_Category_3_max_list']=max_list
test['Product_Category_3_mean_list']=mean_list
test['Product_Category_3_25_list']=twentyfive_list
test['Product_Category_3_75_list']=seventyfive_list
```

```python
from xgboost import XGBClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
```

```python
y=train.Purchase
X=train.drop('Purchase',1)
```

```python

model = XGBClassifier()
learning_rate = [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3]
param_grid = dict(learning_rate=learning_rate)
kfold = StratifiedKFold(y, n_folds=5, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="rmse", n_jobs=-1, cv=kfold)
result = grid_search.fit(X,y)
# summarize results
print("Best: %f using %s" % (result.best_score_, result.best_params_))
means, stdevs = [], []
for params, mean_score, scores in result.grid_scores_:
    stdev = scores.std()
    means.append(mean_score)
    stdevs.append(stdev)
    print("%f (%f) with: %r" % (mean_score, stdev, params))
```
